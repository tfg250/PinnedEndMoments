```python
import plotly.offline as py
import plotly.graph_objs as go
py.init_notebook_mode(connected=True)
import csv
```

```python
csvin = 'txdot_results.csv'
input_file = csv.DictReader(open(csvin))
RESULTS= {}
for row in input_file:
    model = row["Model"]
    RESULTS[model]={k:row[k] for k in row.keys() if k!="Model"}

print(RESULTS['TX_24_100_P01'].keys())
```

```python
x=[float(RESULTS[inp]['span'])/float(RESULTS[inp]['width']) for inp in sorted(RESULTS.keys())]
y=[float(RESULTS[inp]['Y']) for inp in sorted(RESULTS.keys())]
z=[max(abs(float(RESULTS[inp]['maxendmoment'])),abs(float(RESULTS[inp]['minendmoment'])))/float(RESULTS[inp]['maxmoment']) for inp in sorted(RESULTS.keys())]
text = sorted(RESULTS.keys())
trace1 = go.Scatter3d(
    x=x,
    y=y,
    z=z,
    text=text,
    mode='markers',
    hoverinfo='all',
    marker=dict(
        size=5,
        color=z,                # set color to an array/list of desired values
        colorscale='Viridis',   # choose a colorscale
        opacity=0.75
    )
)
data = [trace1]
layout = go.Layout(
                    scene = dict(
                    xaxis = dict(
                        title='Deck Length/Width'),
                    yaxis = dict(
                        title='Girder + Deck Depth (in)'),
                    zaxis = dict(
                        title='End Moment / Max Moment'),),
                    width=700,
                    margin=dict(
                    r=20, b=10,
                    l=10, t=10)
                  )
fig = go.Figure(data=data, layout=layout)
py.iplot(fig)
```
